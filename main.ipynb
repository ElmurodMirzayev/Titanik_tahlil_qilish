**Kutubxonalarni yuklab olish**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Datani yuklab olish**

```python
data = pd.read_csv("Titanic Dataset.csv")
```

**Trik qolganlar tahlili**

```python
all_people = len(data["name"])

alive_people = data[data["survived"] == 1]
alive_female = data[(data["sex"] == "female") & (data["survived"] == 1)]
alive_male = data[(data["sex"] == "male") & (data["survived"] == 1)]


alive_people_len = len(alive_people)

alives = [alive_people_len, all_people-alive_people_len]
labels = ["Survived people", "Could not survived"]

plt.pie(alives, labels=labels, autopct='%1.1f%%')
plt.show()
```

**trik qo'lgan yo'lovchilar qaysi klassdan bo'lganligi**

```python
alive_pclass_1 = data[(data["pclass"] == 1) & (data["survived"] == 1)]
alive_pclass_2 = data[(data["pclass"] == 2) & (data["survived"] == 1)]
alive_pclass_3 = data[(data["pclass"] == 3) & (data["survived"] == 1)]

alive_pclass_1_len = len(alive_pclass_1)
alive_pclass_2_len = len(alive_pclass_2)
alive_pclass_3_len = len(alive_pclass_3)

alives = [alive_pclass_1_len, alive_pclass_2_len, alive_pclass_3_len]
pclasses = ["pclass 1", "pclass 2", "pclass 3"]

plt.bar(pclasses, alives)
plt.show()


```

**Chart diagrammasi yordamida yoshiga qarab trik qolgan odamlarni ko'rish**

```python

alive_teenage = data[(data["age"] < 20) & (data["survived"] == 1)]
alive_between_20_40 = data[(data["age"] >= 20) & (data["age"] < 40) & (data["survived"] == 1)]
alive_bigger_40 = data[(data["age"] >= 40) & (data["survived"] == 1)]

alive_teenage_len = len(alive_teenage)
alive_between_20_40_len = len(alive_between_20_40)
alive_bigger_40_len = len(alive_bigger_40)

alives = [alive_teenage_len, alive_between_20_40_len, alive_bigger_40_len]
ages = ["age < 20", "20 <= age < 40", "40 <= age"]

plt.pie(alives, autopct='%1.1f%%')
plt.legend(ages, loc='upper right', title='Trik qolganlar')
plt.show()
```

**Aka-uka va Opa-singil yaqin tanishlarga qarab tirik qolganlar**

```python
len_sibsp_0_alive = len(data[(data["sibsp"] == 0) & (data["survived"] == 1)])
len_sibsp_1_alive = len(data[(data["sibsp"] == 1) & (data["survived"] == 1)])
len_sibsp_2_alive = len(data[(data["sibsp"] == 2) & (data["survived"] == 1)])
len_sibsp_3_alive = len(data[(data["sibsp"] == 3) & (data["survived"] == 1)])
len_sibsp_4_alive = len(data[(data["sibsp"] == 4) & (data["survived"] == 1)])

values = [len_sibsp_0_alive, len_sibsp_1_alive, len_sibsp_2_alive, len_sibsp_3_alive, len_sibsp_4_alive]
labels = ["Hechkimi yuq", "Yaqin tanishi 1 kishi", "Yaqin tanishi 2 kishi", "Yaqin tanishi 3 kishi", "Yaqin tanishi 4 kishi"]

plt.figure(figsize=(10, 7))
plt.bar(labels, values)
plt.show()

plt.figure(figsize=(9, 9))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.show()
```
