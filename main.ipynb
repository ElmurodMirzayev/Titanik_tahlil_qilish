**Kutubxonalarni yuklab olish**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Datani yuklab olish**

```python
data = pd.read_csv("Titanic Dataset.csv")
```

**3 ta portdan nechta odam trik qolganligini ko'rish**

```python
alive_pclass_1 = data[(data["pclass"] == 1) & (data["survived"] == 1)]
alive_pclass_2 = data[(data["pclass"] == 2) & (data["survived"] == 1)]
alive_pclass_3 = data[(data["pclass"] == 3) & (data["survived"] == 1)]

alive_pclass_1_len = len(alive_pclass_1)
alive_pclass_2_len = len(alive_pclass_2)
alive_pclass_3_len = len(alive_pclass_3)

alives = [alive_pclass_1_len, alive_pclass_2_len, alive_pclass_3_len]
pclasses = ["pclass 1", "pclass 2", "pclass 3"]

plt.bar(pclasses, alives)
plt.show()


```

**Yuqoridagi nechanchi portdagilar ko'proq tirik qolganligini ko'rish**

**-------------------------------------------------------------**

**Chart diagrammasi yordamida yoshiga qarab trik qolgan odamlarni ko'rish**

```python

alive_teenage = data[(data["age"] < 20) & (data["survived"] == 1)]
alive_between_20_40 = data[(data["age"] >= 20) & (data["age"] < 40) & (data["survived"] == 1)]
alive_bigger_40 = data[(data["age"] >= 40) & (data["survived"] == 1)]

alive_teenage_len = len(alive_teenage)
alive_between_20_40_len = len(alive_between_20_40)
alive_bigger_40_len = len(alive_bigger_40)

alives = [alive_teenage_len, alive_between_20_40_len, alive_bigger_40_len]
ages = ["age < 20", "20 <= age < 40", "40 <= age"]

plt.pie(alives, autopct='%1.1f%%')
plt.legend(ages, loc='upper right', title='Trik qolganlar')
plt.show()
```
